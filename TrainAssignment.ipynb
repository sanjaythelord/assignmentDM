# Data collection

```python
import pandas as pd
import sqlite3
# Create a SQL connection to our SQLite database
con = sqlite3.connect("Assignment2024.sqlite")
cur = con.cursor()
```

```python
query = """ select * from train """
# query = """ select * from test """
cur.execute(query)
data = cur.fetchall()
df = pd.DataFrame(data)
df.head(5)
# attribute 31 is the class label containing 0,1 and 2 in train class.
```

# Handling missing values

```python
# function to find which columns have missing data
def missing(df):
  """
  For each attribute/column in the dataframe `df`, count the number of missing entries and return percentage.
  """
  missing_dict = dict()
  total = df.shape[0] # shape[0] is the number of rows
  for attribute in df.columns:
    missing = df[attribute].isna().sum() # count the number of Null/nan/na values
    frac = missing/total * 100 # as a percentage
    missing_dict[attribute] = frac
  return missing_dict
```

```python
m_dict = missing(df)
# m_dict 
# result shows there are no missing data as percentage is 0 
# except for attribute 6 (1.0), 15 (22.36) and 28 (60.160).
```

dropping column where missing value is > 50%

```python
cols_to_drop = [ att for att,frac in m_dict.items() if frac > 50]
cols_to_drop
```

```python
# dropping respective columns
df.drop(columns=cols_to_drop, inplace=True)
# df
```

inserting values in attribute where missing value is < 50% 

```python
# find the columns to impute
cols_to_impute = [ att for att,frac in m_dict.items() if 0 < frac < 50 ]
cols_to_impute
```

```python
# using for incase we have multiple attributes
for col in cols_to_impute:
  mean = df[col].mean() # compute the mean
  # now use the fillna function to replace the NaN avalues with the mean value
  df.fillna(mean, inplace = True)
```
