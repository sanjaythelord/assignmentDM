```python
import numpy as np  
import matplotlib.pyplot as mtp  
import pandas as pd
```

```python
train_data = pd.read_csv(r'trainAssignment.csv',low_memory = False)
test_data = pd.read_csv(r'testAssignment.csv',low_memory = False)
train_data.shape,test_data.shape
```

- .iloc: This method is used for integer-location based indexing in Pandas. It allows you to select rows and columns by their integer positions.
- x_train = train_data.iloc[:, :-1]: This means "select all rows (:) and all columns except the last one (:-1)". The last column is typically assumed to contain the labels (target variable), so this line assigns all feature columns to x_train.
- y_train = train_data.iloc[:, -1] : This line selects all rows (:) but only the last column (-1) from the train_data DataFrame. This is where your labels (the target variable) are located. It assigns these labels to y_train.

```python
x_train = train_data.iloc[:, :-1] # features of train dataset
y_train = train_data.iloc[:, -1] # labels of train dataset
x_test = test_data.iloc[:, :-1] # features of test dataset
y_test = test_data.iloc[:, -1] # labels of test dataset
y_test # it is the one we need to predict
```

# using KNN algorthim

```python
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
```

```python
# Initialize the KNN classifier (choosing k=3)
knn = KNeighborsClassifier(n_neighbors=3) 
# knn_regressor = KNeighborsRegressor(n_neighbors=3)
```

```python
# Train the KNN classifier
knn.fit(x_train, y_train)

# Train the KNN regressor
# knn_regressor.fit(x_train, y_train)
```

```python
# Make predictions on the test set
# y_pred = knn_regressor.predict(x_test)
y_pred_knn = knn.predict(x_test)
```

```python
# printing the predicted value
print(f'Predicted values for y_test: {y_pred_knn}')
```
